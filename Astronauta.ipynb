```python
import pygame, random  # Importa las librerías necesarias
pygame.init()  # Inicia Pygame

# Configuración
ANCHO, ALTO = 1000, 600  # Tamaño de la ventana (ancho más grande para un recorrido más largo)
pantalla = pygame.display.set_mode((ANCHO, ALTO))  # Crea la ventana del juego
pygame.display.set_caption("Astronauta Horizontal")  # Nombre de la ventana

# Colores y reloj
NEGRO, ROJO, VERDE = (0, 0, 0), (255, 0, 0), (0, 255, 0)
reloj = pygame.time.Clock()  # Controla los FPS

# Astronauta (empezando más lejos del lado izquierdo)
astronauta = pygame.Rect(50, 275, 50, 50)  # El astronauta comienza en el lado izquierdo (más lejos de la nave)

# Nave (empezando fuera de la pantalla)
nave = pygame.Rect(1500, 275, 50, 50)  # La nave empieza fuera de la pantalla, en el borde derecho

# Asteroides
asteroides = []

# Tiempo de generación de asteroides
TIEMPO_GENERACION = 50  # Cada 50 cuadros (aproximadamente 1.6 segundos a 30 FPS)
contador_generacion = 0

def crear_asteroide():
    return pygame.Rect(random.randint(900, 950), random.randint(0, ALTO - 50), 50, 50)

def reiniciar():
    global nave
    astronauta.x, astronauta.y = 50, 275  # El astronauta se reinicia al comienzo
    nave.x = 1500  # La nave vuelve a estar fuera de la pantalla
    global asteroides, contador_generacion
    asteroides = []
    contador_generacion = 0

# Bucle principal
jugando = True
while jugando:
    reloj.tick(30)
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            jugando = False

    teclas = pygame.key.get_pressed()
    if teclas[pygame.K_LEFT]: astronauta.x -= 5
    if teclas[pygame.K_RIGHT]: astronauta.x += 5
    if teclas[pygame.K_UP]: astronauta.y -= 5
    if teclas[pygame.K_DOWN]: astronauta.y += 5

    pantalla.fill(NEGRO)
    pygame.draw.rect(pantalla, (255, 255, 255), astronauta)

    # Generación de nuevos asteroides
    contador_generacion += 1
    if contador_generacion >= TIEMPO_GENERACION:
        asteroides.append(crear_asteroide())
        contador_generacion = 0

    # Movimiento de la nave (solo aparecerá cuando el astronauta esté cerca, pero ya no se mueve)
    if astronauta.x > 350 and nave.x == 1500:  # La nave aparece cuando el astronauta ha avanzado
        nave.x = 950  # La nave se posiciona fija en esta posición

    # Si la nave está dentro de la pantalla, la dibujamos
    if nave.x >= 0:
        pygame.draw.rect(pantalla, VERDE, nave)

    # Actualización y dibujo de asteroides
    for ast in asteroides[:]:  # Copia de la lista para evitar problemas al eliminar
        ast.x -= 5  # Movimiento hacia la izquierda
        if ast.x < -50:  # Si el asteroide sale completamente de la pantalla
            asteroides.remove(ast)
        else:
            pygame.draw.rect(pantalla, ROJO, ast)
        if astronauta.colliderect(ast):
            print("¡Perdiste!")
            reiniciar()

    # Comprobación de victoria
    if astronauta.colliderect(nave):
        print("¡Ganaste!")
        reiniciar()

    pygame.display.flip()

pygame.quit()
```
